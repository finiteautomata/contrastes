# Problemas con lugares

En esta notebook, veremos qué pasa con lugares que no aparecen como tales en nuestro listado, pero sí en los de Damián (por ej, "Calahorra").



```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import pandas as pd
import contrastes
import contrastes.geo

df = contrastes.read_occurrence_dataframe("output/provinces_words.csv")
listado_def = pd.read_csv("data/listado_definitivo.csv", index_col=0)
```

```python
lugares = contrastes.geo.places()

df["es_lugar"] = df.index.map(lambda w: w in lugares)
listado_def["es_lugar"] = listado_def["esLugar"] == "lugar"
```

Veamos, de las palabras de nuestro dataframe, cuántas discrepan en si son lugar o no con el del Listado definitivo.

```python
def es_problema(w):
    if (w not in listado_def.index):
        return False
    
    val1 = df.loc[w].es_lugar
    val2 = listado_def.loc[w].es_lugar
    return val1 != val2

df["problema"] = df.index.map(es_problema)

print("Hay {} palabras problemáticas".format(sum(df.problema)))
```
