# DataFrame para un conjunto de usuarios

Haremos la matriz de incidencias para un conjunto de usuarios particular

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
from pymongo import MongoClient

client = MongoClient('localhost', 27017)

db = client['contrastes']
```

```python
users = list(
    db.users.aggregate([{"$sample": {"size": 5000}}], allowDiskUse=True)
)
```

```python
%%time
from contrastes.processing import build_dataframe_from_users

df = build_dataframe_from_users(users)
```

## Generar listado de contrastes/LIW

```python
from contrastes.processing import preprocess_raw_df


df = preprocess_raw_df(df, filter_words=(10, 2))
```

```python
df_without_ival = df.copy()
```

```python
from contrastes.lists import add_ival

add_ival(df, normalize=False)
```

```python
df.sort_values('rank_personas', inplace=True)


for (i, (k, v)) in enumerate(df[:200].iterrows()):
    print("{:<3}. {}".format(i+1, k))
    
```
