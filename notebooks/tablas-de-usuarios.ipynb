Haremos la matriz de incidencias para un conjunto de usuarios particular

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
from pymongo import MongoClient

client = MongoClient('localhost', 27017)

db = client['contrastes']
```

```python
users = list(
    db.users.aggregate([{"$sample": {"size": 4000}}], allowDiskUse=True)
)
```

```python
%%time
from contrastes.processing import build_dataframe_from_users

df = build_dataframe_from_users(users)
```

## Normalizando

En esta notebook normalizaremos las ocurrencias de palabras

```python
df_norm = df.copy()

df_norm.columnas_palabras = df.columnas_palabras
df_norm.columnas_personas = df.columnas_personas
```

```python
df_norm[df.columnas_palabras] = df_norm[df.columnas_palabras] / df_norm[df.columnas_palabras].sum()
```

```python
user_count = {}
for user in users:
    user_count[user["provincia"]] = user_count.get(user["provincia"], 0) + 1

for prov, count in user_count.items():
    df_norm[prov+"_usuarios"] = df_norm[prov+"_usuarios"] / count
```
