```python
import glob
import os
path = "/media/jmperez/Ramiro/jmperez/tesis/tweets/"

directorios = glob.glob(os.path.join(path, "**/"))
files_to_use = ["00{}.json".format(i) for i in range(1,7)]

db.tweets.drop()
db.tweets.create_index([('id', pymongo.ASCENDING)], unique=True)
db.tweets.create_index([('text', pymongo.TEXT)], default_language='spanish')

print(directorios)
print(files_to_use)

def serialize(tweet, provincia):
    ret = {k:tweet[k] for k in ["id", "text", "place"]}
    ret["provincia"] = provincia
    
    return ret

for path_to_province in directorios:
    provincia = path_to_province.split("/")[-2]
    print("="*80)
    print(provincia)
    files = [os.path.join(path_to_province, file) for file in files_to_use]
    
    for file in files:
        print(file)
        tweets = json.load(open(file))
        tweets = [serialize(tweet, provincia) for tweet in tweets]
        
        db.tweets.insert_many(tweets)
```
