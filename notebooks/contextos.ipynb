```python
import pymongo
import pickle

database_name = "contrastes"

client = pymongo.MongoClient('localhost', 27017)
db = client[database_name]
```

```python
import json

tweets = json.load(open("data/sample_tweets.json"))
```

```python
tweets[0].keys()
```

```python
search = "maradona"
for tweet in db.tweets.find({"$text": {"$search": search}}):
    print("="*40)
    print(tweet["text"])
```
