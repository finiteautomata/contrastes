# Contextos 

```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import pymongo

client = pymongo.MongoClient("localhost", 27017)
db = client["contrastes"]

df = pd.read_csv("../output/listados/Listado a Etiquetar 2019.csv", index_col=0)
```

Para cada palabra, daremos contextos

```python
for word in df.index:
    print("="*80)
    print(f"Contextos de {word}\n")
    context_tweets = db.tweets.find({"$text": {"$search": word}}).limit(15)
    
    for tweet in context_tweets: 
        print(f'-- {tweet["text"]}')
    print("\n")
```
