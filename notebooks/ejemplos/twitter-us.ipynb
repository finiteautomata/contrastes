# Twitter-US

En esta notebook, exploraremos los datos de [Roller 2012](https://dollar.biz.uiowa.edu/~street/p1500-roller.pdf)


```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import pandas as pd

df_train_small = pd.read_table("data/twitter-us/training-small.txt", header=None,
                               names=["username", "timestamp", "lat", "lon", 
                               "#followers", "#following", "one", "blank", "text"])
```

```python

df_dev = pd.read_table("data/twitter-us/dev.txt", header=None,
                        names=["username", "timestamp", "lat", "lon", 
                               "#followers", "#following", "one", "blank", "text"])

df_test = pd.read_table("data/twitter-us/test.txt", header=None,
                        names=["username", "timestamp", "lat", "lon", 
                            "#followers", "#following", "one", "blank", "text"])
```

```python
print("Cantidad de instancias de train-small: {:<3.2f}M instancias".format(len(df_train_small)/1e6))
print("Cantidad de instancias de dev:         {:<3.2f}M instancias".format(len(df_dev)/1e6))
print("Cantidad de instancias de test:        {:<3.2f}M instancias".format(len(df_test)/1e6))

```
